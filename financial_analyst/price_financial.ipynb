```python
import os
import sys


ROOT_CRAWL = os.path.join(os.path.dirname("__file__"), "..", "crawl")
```

# Load data

```python
import json

with open(os.path.join(ROOT_CRAWL, 'list_stocks.json'), 'r') as f:
    list_stocks = json.load(f)
list_stocks
```

```python
import pandas as pd

balancesheet = pd.read_csv(os.path.join(ROOT_CRAWL, "balancesheet.csv"), encoding="utf-8")
balancesheet["quarter_start_time"] = balancesheet.apply(
    lambda row: (pd.Period(freq="Q", year=row["year"], quarter=row["quarter"]) + 1).start_time,
    axis=1,
)
balancesheet["quarter_end_time"] = balancesheet.apply(
    lambda row: (pd.Period(freq="Q", year=row["year"], quarter=row["quarter"]) + 1).end_time, axis=1
)
balancesheet
```

```python
financialratio = pd.read_csv(os.path.join(ROOT_CRAWL, "financialratio.csv"), encoding="utf-8")
financialratio["quarter_start_time"] = financialratio.apply(
    lambda row: (pd.Period(freq="Q", year=row["year"], quarter=row["quarter"]) + 1).start_time,
    axis=1,
)
financialratio["quarter_end_time"] = financialratio.apply(
    lambda row: (pd.Period(freq="Q", year=row["year"], quarter=row["quarter"]) + 1).end_time, axis=1
)
financialratio
```

```python
incomestatement = pd.read_csv(os.path.join(ROOT_CRAWL, "incomestatement.csv"), encoding="utf-8")
incomestatement["quarter_start_time"] = incomestatement.apply(
    lambda row: (pd.Period(freq="Q", year=row["year"], quarter=row["quarter"]) + 1).start_time,
    axis=1,
)
incomestatement["quarter_end_time"] = incomestatement.apply(
    lambda row: (pd.Period(freq="Q", year=row["year"], quarter=row["quarter"]) + 1).end_time, axis=1
)
incomestatement.info()
```

```python
prices = pd.read_csv(os.path.join(ROOT_CRAWL, "prices.csv"), encoding="utf-8")
prices["date"] = pd.to_datetime(prices["date"])
prices
```

# Price and financialratio

```python
[x["CategoryName"] for x in list_stocks["category"]]
```

```python
# symbols = list(filter(lambda x: x["CategoryName"] == "Ngân Hàng", list_stocks["category"]))[0][
#     "ListStockCode"
# ]
symbols = ["HPG"]
assert len(symbols) == 1
```

```python
price_financialratio = pd.merge(
    prices[prices["symbol"].isin(symbols)],
    financialratio[financialratio["ticker"].isin(symbols)],
    left_on="symbol",
    right_on="ticker",
    how="inner",
)
price_financialratio = price_financialratio[
    (price_financialratio["date"] >= price_financialratio["quarter_start_time"])
    & (price_financialratio["date"] <= price_financialratio["quarter_end_time"])
]
if len(symbols) == 1:
    price_financialratio = price_financialratio.drop_duplicates(subset="date", keep="first")
price_financialratio.info()
```

## P/E & P/B

```python
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go


price_financialratio["pe"] = (
    price_financialratio["close"] * 1000 / price_financialratio["earningPerShare"]
)
# sns.lineplot(data=price_financialratio, x="date", y="pe", hue="symbol")
fig = px.line(price_financialratio, x="date", y="pe", color="symbol")
# fig.update_layout(yaxis_range=[0,100])
fig.show()
```

```python
if len(symbols) == 1:
    print(
        price_financialratio[
            (price_financialratio["date"] > "2022-07-01")
            & (price_financialratio["date"] < "2022-12-01")
        ]["pe"].min()
    )
```

```python
price_financialratio["pb"] = (
    price_financialratio["close"] * 1000 / price_financialratio["bookValuePerShare"]
)
# sns.lineplot(data=price_financialratio, x="date", y="pb", hue="symbol")
fig = px.line(price_financialratio, x="date", y="pb", color='symbol')
fig.show()
```

```python
if len(symbols) == 1:
    fair_pb = 1.5
    filtered_date = price_financialratio[
        (price_financialratio["date"] > "2022-07-01")
        & (price_financialratio["date"] < "2022-12-01")
    ]
    row = filtered_date.loc[filtered_date["pb"].idxmin()]

    print("Giá fair:", ((fair_pb) / row["pb"]) * row["close"])
    print("Tăng % từ đáy: ", ((fair_pb - row["pb"]) / row["pb"]) * 100)
```
