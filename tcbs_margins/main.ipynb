```python
def to_number(x):
    x = x.replace(',', '.')
    try:
        return float(x)
    except:
        return x
```

```python
with open('data_7_11_2022.csv', 'r') as f:
    data_7_11_2022 = [list(map(to_number, x.split())) for x in f.readlines()]
data_7_11_2022 = dict(map(lambda x: (x[1], x), data_7_11_2022))
data_7_11_2022
```

```python
with open('data_14_11_2022.csv', 'r') as f:
    data_14_11_2022 = [list(map(to_number, x.split())) for x in f.readlines()]
data_14_11_2022 = dict(map(lambda x: (x[1], x), data_14_11_2022))
data_14_11_2022
```

```python
with open('data_18_11_2022.csv', 'r') as f:
    data_18_11_2022 = [list(map(to_number, x.split())) for x in f.readlines()]
data_18_11_2022 = dict(map(lambda x: (x[1], x), data_18_11_2022))
data_18_11_2022
```

```python
factor = 0.0
```

```python
for code in set(list(data_14_11_2022.keys()) + list(data_7_11_2022.keys())):
    if (
        (code in set(data_14_11_2022.keys()) and code in set(data_7_11_2022.keys()))
    ):
        price_before = data_7_11_2022[code][4]
        price_after = data_14_11_2022[code][4]
        if abs(price_after - price_before) > factor * price_before:
            print(code, price_before, price_after)

for code in set(list(data_14_11_2022.keys()) + list(data_7_11_2022.keys())):
    if (
        (code in set(data_14_11_2022.keys()) and code not in set(data_7_11_2022.keys()))
        or (code not in set(data_14_11_2022.keys()) and code in set(data_7_11_2022.keys()))
    ):
        print(code)
```
